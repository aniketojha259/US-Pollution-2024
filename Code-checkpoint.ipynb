### 8. Air Quality Index (AQI) Classification:
Can we classify the air quality based on the Air Quality Index (AQI) for the pollutants? Use AQI values for O3, CO, SO2, and NO2 to classify air quality into categories like "Good", "Moderate", "Unhealthy for Sensitive Groups", "Unhealthy", "Very Unhealthy", and "Hazardous".

### Step 1: Define AQI Categories
We can define AQI categories based on the EPA's (Environmental Protection Agency) standards.

```python
def classify_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazardous'
```

### Step 2: Apply Classification to Each Pollutant's AQI
We will apply the classify_aqi function to the AQI columns for Ozone (O3 AQI), Carbon Monoxide (CO AQI), Sulphur Dioxide (SO2 AQI), and Nitrogen Dioxide (NO2 AQI).

```python
# Apply the classify_aqi function to each AQI column
df1['O3_AQI_Category'] = df1['O3 AQI'].apply(classify_aqi)
df1['CO_AQI_Category'] = df1['CO AQI'].apply(classify_aqi)
df1['SO2_AQI_Category'] = df1['SO2 AQI'].apply(classify_aqi)
df1['NO2_AQI_Category'] = df1['NO2 AQI'].apply(classify_aqi)

# Display the new columns
df1[['O3 AQI', 'O3_AQI_Category', 'CO AQI', 'CO_AQI_Category', 'SO2 AQI', 'SO2_AQI_Category', 'NO2 AQI', 'NO2_AQI_Category']].head()
```

### Step 3: Visualize AQI Classifications
We can create a bar plot or count plot to visualize the frequency of each AQI category for each pollutant.

```python
# Plot AQI categories for Ozone (O3)
plt.figure(figsize=(10,6))
sns.countplot(x='O3_AQI_Category', data=df1, order=['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous'])
plt.title('Ozone (O3) AQI Categories')
plt.show()

# Repeat similar plots for CO, SO2, and NO2 AQI categories
```

### 11. Identify Outliers
Q: Are there any outliers in pollutant levels? What methods can you use to detect them?

Solution:

Use the Interquartile Range (IQR) method to identify outliers.
Calculate the IQR for each pollutant and filter out the outliers.

```python
# Function to detect outliers using IQR
def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return data[(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))]

# Check for outliers in O3 Mean
o3_outliers = detect_outliers(df['O3 Mean'])
print("O3 Outliers:")
print(o3_outliers)
```
